```python
import pandas as pd 
import numpy as np
```

```python
# è¯»å–onlineæ•°æ®ï¼Œä¹‹å‰çš„ç‰¹å¾å·¥ç¨‹å¥½åƒæ²¡åšå®Œå…¨ï¼Œç»§ç»­åš
df_online = pd.read_csv('./df_online.csv')
df_online
```

```python
# è¯»å–offlineæ•°æ®ï¼Œä¹‹å‰çš„ç‰¹å¾å·¥ç¨‹å¥½åƒæ²¡åšå®Œå…¨ï¼Œç»§ç»­åš
df_offline = pd.read_csv('./df_offline.csv')
df_offline
```

```python
# ä¿®æ”¹è·ç¦»ä¸ºæ­£å¸¸æ•°å€¼
def change_dist(x):
    if x < 0:
        return 0
    elif x == 0:
        return 500
    elif x > 10:
        return 5000
    else:
        return x*500
```

```python
df_offline['Distance'] = df_offline['Distance'].apply(change_dist)
```

```python
# æ»¡å¤šå°‘ä¼˜æƒ çš„ä¿®æ”¹
ser01 = df_offline['discount_rate_x'].replace(-1,0)
df_offline['discount_rate_x'] = ser01
ser02 = df_offline['discount_rate_y'].replace(-10,0)
df_offline['discount_rate_y'] = ser02
```

```python
# æ‰“æŠ˜çš„ä¿®æ”¹
def discount_rate_ratio(x):
    if x == -10.0:
        return 0
    else:
        return 1 - x
```

```python
df_offline['discount_rate_ratio'] = df_offline['discount_rate_ratio'].apply(discount_rate_ratio)
```

```python
# æ ·æœ¬ç±»åˆ«æžé”™äº†
def chang_sample(x):
    if x == 0:
        return 2
    elif x == 1:
        return 1
    elif x == 2:
        return 0
df_offline['samples_class'] = df_offline['samples_class'].apply(chang_sample)
```

```python
# èŽ·å–æ­£è´Ÿæ ·æœ¬
df_offline1 = df_offline[list(df_offline.iloc[:,-1] != 2)]
df_offline1
```

```python
df_offline2 = df_offline1.drop(['use_year','use_month','use_day','get_year','get_month','get_day'], axis = 1)
df_offline2
```

```python
# è¯»å–offline_predictæ•°æ®ï¼Œä¹‹å‰çš„ç‰¹å¾å·¥ç¨‹å¥½åƒæ²¡åšå®Œå…¨ï¼Œç»§ç»­åš
df_offline_predict = pd.read_csv('./df_offline_predict.csv')
df_offline_predict1 = df_offline_predict[list(df_offline2.columns)[:-1]]
df_offline_predict1['Date_received'] = df_offline_predict['Date_received']
df_offline_predict1['discount_rate_ratio'] = df_offline_predict1['discount_rate_ratio'].apply(discount_rate_ratio)
df_offline_predict1['Distance'] = df_offline_predict1['Distance'].apply(change_dist)
ser01 = df_offline_predict1['discount_rate_x'].replace(-1,0)
df_offline_predict1['discount_rate_x'] = ser01
ser02 = df_offline_predict1['discount_rate_y'].replace(-10,0)
df_offline_predict1['discount_rate_y'] = ser02
df_offline_predict1
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import precision_score,accuracy_score,recall_score,f1_score,mean_squared_error
import xgboost as xgb
```

```python
# df_offline2[list(df_offline2)[2:7]]
df_offline2[list(df_offline2)[-1]]
```

```python
# æ•°æ®åˆ’åˆ†
x_train,x_test,y_train,y_test = train_test_split(df_offline2[list(df_offline2)[2:7]],
                                df_offline2[list(df_offline2)[-1]], test_size = 0.2,
                                random_state = 28)
print('è®­ç»ƒæ ·æœ¬æ•°é‡ï¼š{}ï¼Œæµ‹è¯•æ ·æœ¬æ•°é‡ï¼š{}'.format(x_train.shape[0], x_test.shape[0]))
```

```python
# å°†æ•°æ®è½¬ä¸ºXGBoostå¯ç”¨çš„æ•°æ®
dtrain = xgb.DMatrix(x_train, label = y_train)
dtest = xgb.DMatrix(x_test)
# xtest2 = 
```

```python
# XGBoostæž„å»ºæ¨¡åž‹
# 1ï¼Œå‚æ•°è®¾å®š
params={
'booster':'gbtree', #gbtree:åŸºäºŽæ ‘æ¨¡åž‹ï¼Œgbliner:åŸºäºŽçº¿æ€§æ¨¡åž‹
# 'objective':'multi:softmax', #è¿”å›žçš„æ˜¯ç±»åˆ«ï¼Œéœ€è¦è®¾ç½®num_class 
# 'booster':'gblinear',
'objective': 'reg:logistic', #äºŒåˆ†ç±»çš„é—®é¢˜ï¼Œè¿”å›žæ¦‚çŽ‡
# 'num_class':2, # åˆ†ç±»ä¸ªæ•°
'gamma':0.1,  # ç”¨äºŽæŽ§åˆ¶æ˜¯å¦åŽå‰ªæžçš„å‚æ•°,è¶Šå¤§è¶Šä¿å®ˆï¼Œä¸€èˆ¬0.1ã€0.2è¿™æ ·å­ã€‚
'max_depth':7, # æž„å»ºæ ‘çš„æ·±åº¦ï¼Œè¶Šå¤§è¶Šå®¹æ˜“è¿‡æ‹Ÿåˆ
# 'alpha': L1æƒé‡çš„è®¾ç½®
'lambda':4,  # æŽ§åˆ¶æ¨¡åž‹å¤æ‚åº¦çš„æƒé‡å€¼çš„L2æ­£åˆ™åŒ–é¡¹å‚æ•°ï¼Œå‚æ•°è¶Šå¤§ï¼Œæ¨¡åž‹è¶Šä¸å®¹æ˜“è¿‡æ‹Ÿåˆã€‚
'subsample':0.8, # éšæœºé‡‡æ ·è®­ç»ƒæ ·æœ¬
'colsample_bytree':0.8, # ç”Ÿæˆæ ‘æ—¶è¿›è¡Œçš„åˆ—é‡‡æ ·
'min_child_weight':3, #æœ€å°å¶å­ç»“ç‚¹çš„æƒé‡å’Œ
'eval_metric':'auc', 
# 'max_leaf_nodes':10 #æ ‘ä¸Šæœ€å¤§çš„å¶å­ç»“ç‚¹æˆ–å¶å­çš„æ•°é‡
# è¿™ä¸ªå‚æ•°é»˜è®¤æ˜¯ 1ï¼Œæ˜¯æ¯ä¸ªå¶å­é‡Œé¢ h çš„å’Œè‡³å°‘æ˜¯å¤šå°‘ï¼Œå¯¹æ­£è´Ÿæ ·æœ¬ä¸å‡è¡¡æ—¶çš„ 0-1 åˆ†ç±»è€Œè¨€
#ï¼Œå‡è®¾ h åœ¨ 0.01 é™„è¿‘ï¼Œmin_child_weight ä¸º 1 æ„å‘³ç€å¶å­èŠ‚ç‚¹ä¸­æœ€å°‘éœ€è¦åŒ…å« 100 ä¸ªæ ·æœ¬ã€‚
#è¿™ä¸ªå‚æ•°éžå¸¸å½±å“ç»“æžœï¼ŒæŽ§åˆ¶å¶å­èŠ‚ç‚¹ä¸­äºŒé˜¶å¯¼çš„å’Œçš„æœ€å°å€¼ï¼Œè¯¥å‚æ•°å€¼è¶Šå°ï¼Œè¶Šå®¹æ˜“ overfittingã€‚
'silent':0 ,#è®¾ç½®æˆ1åˆ™æ²¡æœ‰è¿è¡Œä¿¡æ¯è¾“å‡ºï¼Œæœ€å¥½æ˜¯è®¾ç½®ä¸º0.
'eta': 0.001, # å¦‚åŒå­¦ä¹ çŽ‡
'seed':28,
'pred_margin':0,
# 'max_delta_step':0 é™åˆ¶æ¯æ£µæ ‘æƒé‡æ”¹å˜çš„æœ€å¤§æ­¥é•¿
# 'scale_pos_weight' :1 ,åœ¨å„ç±»åˆ«æ ·æœ¬ååˆ†ä¸å¹³è¡¡æ—¶ï¼ŒæŠŠè¿™ä¸ªå‚æ•°è®¾å®šä¸ºä¸€ä¸ªæ­£å€¼ï¼Œå¯ä»¥ä½¿ç®—æ³•æ›´å¿«æ”¶æ•›ã€‚
# 'nthread':7,# cpu çº¿ç¨‹æ•°
#'eval_metric': 'auc'
}
```

```python
num_round = 5
pred_margin = 3.0
# 2ï¼Œæ¨¡åž‹è®­ç»ƒ
'''
å‘½ä»¤è¡Œå‚æ•°
ä»¥ä¸‹å‚æ•°ä»…ç”¨äºŽæŽ§åˆ¶å°ç‰ˆæœ¬çš„xgboost
1.use_bufferé»˜è®¤å€¼ä¸º1ï¼ŒæŽ§åˆ¶æ˜¯å¦ä»Žæ–‡æœ¬è¾“å‡ºåˆ›å»ºåŒç¼“å†²åŒºï¼Œé€šè¿‡æ­¤å¯ä»¥åŠ é€ŸåŠ è½½æ—¶é—´
2.num_roundè¡¨ç¤ºboostingçš„å¾ªçŽ¯æ¬¡æ•°
3.dataè¡¨ç¤ºè®­ç»ƒæ•°æ®çš„è·¯å¾„
4.test:dataè¡¨ç¤ºç”¨äºŽé¢„æµ‹çš„æµ‹è¯•æ•°æ®é›†çš„è·¯å¾„
5.save_periodé»˜è®¤å€¼ä¸º0ï¼Œè¡¨ç¤ºæ¨¡åž‹ä¿å­˜çš„å‘¨æœŸæ•°
è¯¥å€¼è®¾ä¸º10è¡¨ç¤ºæ¯è¿›è¡Œ10æ¬¡xgboostå¾ªçŽ¯å°±ä¿å­˜ä¸€æ¬¡æ¨¡åž‹ï¼Œè®¾ä¸º0è¡¨ç¤ºè®­ç»ƒè¿‡ç¨‹ä¸­ä¸ä¿å­˜ä»»ä½•æ¨¡åž‹
6.taské»˜è®¤å€¼ä¸ºtrainï¼Œå¯å–å€¼å¦‚ä¸‹ï¼š
train:ä½¿ç”¨dataè¿›è¡Œè®­ç»ƒ
pred:ä½¿ç”¨test:dataè¿›è¡Œé¢„æµ‹
eval:è®¡ç®—ç»Ÿè®¡é‡çš„å€¼ï¼Œç»Ÿè®¡é‡é€šè¿‡eval[name]=filename
dump:æŠŠå­¦ä¹ å¥½çš„æ¨¡åž‹å†™æˆæ–‡æœ¬æ ¼å¼
7.model_iné»˜è®¤å€¼ä¸ºNULLï¼Œè¡¨ç¤ºè¾“å…¥æ¨¡åž‹æ‰€åœ¨çš„è·¯å¾„
8.model_outé»˜è®¤å€¼ä¸ºNULLï¼Œè¡¨ç¤ºè®­ç»ƒç»“æŸåŽè¾“å‡ºæ¨¡åž‹æ‰€åœ¨çš„è·¯å¾„
9.model_dirè¡¨ç¤ºè®­ç»ƒè¿‡ç¨‹ä¸­ä¿å­˜çš„æ¨¡åž‹çš„è¾“å‡ºç›®å½•
10.fmap:feature map, used for dump model
11.name_dumpé»˜è®¤å€¼ä¸ºdump.txtï¼Œè¡¨ç¤ºè¾“å‡ºçš„æ¨¡åž‹æ‰€åœ¨æ–‡ä»¶çš„åå­—
12.name_predé»˜è®¤å€¼ä¸ºpred.txtï¼Œè¡¨ç¤ºé¢„æµ‹æ–‡ä»¶çš„åå­—
13.pred_marginé»˜è®¤å€¼ä¸º0ï¼Œè¡¨ç¤ºpredict marginè€Œä¸æ˜¯è½¬æ¢çš„æ¦‚çŽ‡'''
bst = xgb.train(params, dtrain, num_round)
# 3ï¼Œæ¨¡åž‹ä¿å­˜
bst.save_model('xgb.model')
```

```python
# XGBoostæ¨¡åž‹é¢„æµ‹
y_pred = bst.predict(dtest)
print(mean_squared_error(y_pred, y_test))
dtrain1 = xgb.DMatrix(x_train)
y_pred1 = bst.predict(dtrain1)
print(y_pred1)
# ss01 = pd.Series(y_pred1)
# ss01.to_csv('ss01.csv',index = False)
y_pred3 = change_ratio(y_pred1)
print("è®­ç»ƒé›†å‡†ç¡®çŽ‡ä¸º:{}%" .format(accuracy_score(y_train, y_pred3)*100))
```

```python
# æ¦‚çŽ‡è½¬æ¢ä¸ºç±»åˆ«,éœ€æ±‚è½¬æ¢æ–¹ä¾¿
def change_ratio(x):
    if x[0] == 1 or x[0] == 0:
        return x
    else:
        return np.where(x >= 0.5, 1, 0)
```

```python
# 4.åŠ è½½æ¨¡åž‹
bst2 = xgb.Booster()
bst2.load_model('xgb.model')
# 5ï¼Œä½¿ç”¨åŠ è½½æ¨¡åž‹è¿›è¡Œé¢„æµ‹
y_pred2 = bst2.predict(dtest)
print(mean_squared_error(y_pred2, y_test))
print(y_pred2)
y_pred3 = change_ratio(y_pred2)
y_pred3
```
